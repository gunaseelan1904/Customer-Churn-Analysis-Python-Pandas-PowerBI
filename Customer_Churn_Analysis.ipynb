# Part 3: Exploratory Data Analysis

## Objective

The objective of this stage is to analyze the cleaned Telco Customer Churn dataset to identify customer behavior patterns, measure churn across different customer segments, and generate actionable business insights that support data-driven decision-making.

## 3.1 Overall Customer Distribution

The distribution of retained and churned customers is examined to understand the overall composition of the customer base before performing detailed churn analysis.

```python
print("Total Customers:", len(df))

print("\nCustomer Distribution:")

print(df["Churn"].value_counts())
```

### Business Insight

The cleaned dataset contains **7,032 customers**, of which **5,163 customers (73.4%)** were retained and **1,869 customers (26.6%)** had churned. The dataset is moderately imbalanced, with retained customers forming the majority, which is an important consideration when interpreting churn patterns.

## 3.2 Calculate Overall Churn Rate

The overall churn rate is calculated to measure the percentage of customers who discontinued the service. This serves as a key business metric for evaluating customer retention.

```python
churn_rate = (df["Churn"] == "Yes").mean() * 100

print(f"Overall Churn Rate: {churn_rate:.2f}%")
```

### Business Insight

The overall customer churn rate is **26.58%**, indicating that approximately **1 in every 4 customers** discontinued the service. This suggests a relatively high customer attrition rate and highlights the need to identify the key factors contributing to customer churn.

## 3.3 Churn Rate by Gender

The churn rate is calculated for male and female customers to determine whether gender has a meaningful relationship with customer retention.

```python
gender_churn_rate = (
    pd.crosstab(df["gender"], df["Churn"], normalize="index") * 100
).round(2)

print(gender_churn_rate)
```

### Business Insight

Female customers have a slightly higher churn rate (**26.96%**) than male customers (**26.20%**). However, the difference is less than **1 percentage point**, indicating that **gender has minimal influence on customer churn** and is unlikely to be a primary driver of customer attrition.

## 3.4 Churn Rate by Senior Citizen Status

The churn rate is calculated for senior citizens and non-senior citizens to determine whether age group influences customer retention.

```python
senior_churn_rate = (
    pd.crosstab(df["SeniorCitizen"], df["Churn"], normalize="index") * 100
).round(2)

senior_churn_rate.index = ["Non-Senior", "Senior Citizen"]

print(senior_churn_rate)
```

### Business Insight

Senior citizens have a significantly higher churn rate (**41.68%**) compared to non-senior customers (**23.65%**). This suggests that senior citizens are considerably more likely to discontinue the service and may require targeted retention strategies, such as personalized customer support, loyalty programs, or service plans tailored to their needs.

## 3.5 Churn Rate by Contract Type

The churn rate is calculated for each contract type to determine how customer commitment influences customer retention.

```python
contract_churn_rate = (
    pd.crosstab(df["Contract"], df["Churn"], normalize="index") * 100
).round(2)

print(contract_churn_rate)
```

### Business Insight

Customers with **month-to-month contracts** have the highest churn rate (**42.71%**), whereas customers with **one-year contracts** have a much lower churn rate (**11.28%**). Customers with **two-year contracts** exhibit the lowest churn rate (**2.85%**). This indicates that **long-term contracts are strongly associated with higher customer retention**, making contract commitment one of the most influential factors affecting customer churn.

## 3.6 Churn Rate by Payment Method

The churn rate is calculated for each payment method to determine whether customers using different payment options exhibit different retention patterns.

```python
payment_churn_rate = (
    pd.crosstab(df["PaymentMethod"], df["Churn"], normalize="index") * 100
).round(2)

print(payment_churn_rate)
```

### Business Insight

Customers using **Electronic Check** have the highest churn rate (**45.29%**), which is significantly higher than all other payment methods. In contrast, customers using **Credit Card (automatic)** (**15.25%**) and **Bank Transfer (automatic)** (**16.73%**) have the lowest churn rates. This suggests that customers enrolled in **automatic payment methods are more likely to remain with the company**, while Electronic Check users represent a high-risk customer segment requiring targeted retention efforts.

## 3.7 Churn Rate by Internet Service

The churn rate is calculated for each internet service type to determine whether the type of internet service influences customer retention.

```python
internet_churn_rate = (
    pd.crosstab(df["InternetService"], df["Churn"], normalize="index") * 100
).round(2)

print(internet_churn_rate)
```

### Business Insight

Customers using **Fiber Optic** internet service have the highest churn rate (**41.89%**), which is more than double the churn rate of **DSL** customers (**19.00%**) and significantly higher than customers with **no internet service** (**7.43%**). This suggests that customers subscribed to Fiber Optic services are at a much greater risk of churning, indicating a need to investigate factors such as pricing, service quality, or customer satisfaction within this segment.

## 3.8 Churn Rate by Tenure Group

Customers are grouped based on their tenure to examine how the length of their relationship with the company influences customer churn.

```python
df["TenureGroup"] = pd.cut(
    df["tenure"],
    bins=[0, 12, 24, 48, 72],
    labels=["0-12 Months", "13-24 Months", "25-48 Months", "49-72 Months"],
    include_lowest=True
)

tenure_churn_rate = (
    pd.crosstab(df["TenureGroup"], df["Churn"], normalize="index") * 100
).round(2)

print(tenure_churn_rate)
```

### Business Insight

Customers in the **0–12 months** tenure group have the highest churn rate (**47.68%**), indicating that nearly half of new customers discontinue the service within their first year. The churn rate steadily decreases as customer tenure increases, reaching only **9.51%** for customers with **49–72 months** of tenure. This suggests that customer loyalty strengthens over time, making the first year a critical period for customer retention initiatives.

## 3.9 Average Monthly Charges by Churn Status

The average monthly charges are compared between retained and churned customers to determine whether pricing is associated with customer churn.

```python
monthly_charge = (
    df.groupby("Churn")["MonthlyCharges"]
      .mean()
      .round(2)
)

print(monthly_charge)
```

### Business Insight

Churned customers have a higher average monthly charge (**74.44**) than retained customers (**61.31**). This suggests that customers with higher monthly charges are more likely to discontinue the service, indicating that pricing or the perceived value of premium plans may influence customer retention.

## 3.10 Correlation Analysis

The relationship between numerical variables is examined to understand how customer tenure, monthly charges, and total charges are associated with one another.

```python
correlation_matrix = df[["tenure", "MonthlyCharges", "TotalCharges"]].corr()

print(correlation_matrix.round(2))
```

### Business Insight

A **strong positive correlation** exists between **tenure** and **TotalCharges** (**0.83**), indicating that customers who remain with the company longer tend to accumulate higher total charges. **MonthlyCharges** also show a **moderate positive correlation** with **TotalCharges** (**0.65**), while the relationship between **MonthlyCharges** and **tenure** is relatively weak (**0.25**). This suggests that total customer value is driven primarily by customer retention rather than higher monthly pricing alone.

# Part 3 Summary

### Key Findings

- The overall customer churn rate is **26.58%**, indicating that approximately **1 in every 4 customers** discontinued the service.
- Gender has minimal influence on customer churn, with only a small difference in churn rates between male and female customers.
- Senior citizens have a significantly higher churn rate than non-senior customers, making them a higher-risk customer segment.
- Customers with **month-to-month contracts** exhibit the highest churn rate, while **two-year contracts** have the lowest churn rate, highlighting contract duration as a major factor influencing customer retention.
- Customers using **Electronic Check** have the highest churn rate, whereas customers using automatic payment methods experience substantially lower churn.
- **Fiber Optic** customers have the highest churn rate among internet service types, indicating the need for further investigation into service quality, pricing, or customer satisfaction.
- Customers within their **first year of service** are the most likely to churn, while long-term customers demonstrate significantly higher retention.
- Churned customers have higher average monthly charges than retained customers, suggesting that pricing or the perceived value of premium plans may influence customer decisions.
- Correlation analysis indicates that customer lifetime value is strongly associated with tenure, emphasizing the importance of long-term customer retention strategies.

# Part 4: Feature Engineering and Data Preparation

## Objective

The objective of this stage is to create additional analytical features that improve reporting and visualization in Power BI. These derived columns simplify business analysis and support the development of interactive dashboards.

## 4.1 Create Customer Status

A customer status column is created to provide more meaningful labels for reporting and dashboard visualizations.

```python
df["CustomerStatus"] = df["Churn"].replace({
    "Yes": "Churned",
    "No": "Retained"
})

print(df[["Churn", "CustomerStatus"]].head())
```

### Observation

A new **CustomerStatus** column was successfully created by replacing the original **Churn** values with more descriptive labels (**Retained** and **Churned**). This improves readability and makes dashboard reporting easier to understand.

## 4.2 Create Tenure Category

Customers are grouped into tenure categories to simplify retention analysis across different stages of the customer lifecycle.

```python
df["TenureCategory"] = pd.cut(
    df["tenure"],
    bins=[0, 12, 24, 48, 72],
    labels=[
        "0-12 Months",
        "13-24 Months",
        "25-48 Months",
        "49-72 Months"
    ],
    include_lowest=True
)

print(df["TenureCategory"].value_counts())
```

### Observation

The **TenureCategory** column was successfully created by grouping customers into four tenure ranges based on their length of service. This derived feature simplifies customer lifecycle analysis and supports more effective reporting and visualization in Power BI.

## 4.3 Create Monthly Charge Category

Monthly charges are grouped into pricing bands to simplify customer pricing analysis.

```python
df["MonthlyChargeCategory"] = pd.cut(
    df["MonthlyCharges"],
    bins=[0, 35, 70, 120],
    labels=[
        "Low",
        "Medium",
        "High"
    ]
)

print(df["MonthlyChargeCategory"].value_counts())
```

### Observation

The **MonthlyChargeCategory** column was successfully created by grouping customers into **Low**, **Medium**, and **High** monthly charge bands. This derived feature simplifies pricing analysis and supports clearer segmentation and visualization in the Power BI dashboard.

## 4.4 Calculate Revenue at Risk

Revenue at Risk is calculated by summing the monthly charges of churned customers. This metric estimates the recurring monthly revenue lost due to customer churn.

```python
revenue_at_risk = df.loc[
    df["Churn"] == "Yes",
    "MonthlyCharges"
].sum()

print(f"Monthly Revenue at Risk: {revenue_at_risk:,.2f}")
```

### Observation

The **Monthly Revenue at Risk** was successfully calculated as **139130.85**, representing the estimated recurring monthly revenue associated with customers who have churned. This metric can be used as a key performance indicator (KPI) in the Power BI dashboard to monitor the financial impact of customer churn.

### Observation

The newly created features (**CustomerStatus**, **TenureCategory**, and **MonthlyChargeCategory**) were successfully verified in the dataset. The preview confirms that the derived columns have been added correctly and are ready to support analysis and visualization in Power BI.
